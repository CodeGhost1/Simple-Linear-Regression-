```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv("C:/Users/Dell/Desktop/EXCELR/Assignment - Data Science/SLR/Salary_Data.csv")
df.head()
```

```python
df1 = df.rename({"YearsExperience":"y_exp", "Salary":"sal"}, axis=1)
```

```python
df1.info()
```

```python
df1.corr()
```

```python
sns.distplot(df1.y_exp)
```

```python
sns.distplot(df1.sal)
```

```python
import statsmodels.formula.api as smf
model = smf.ols("sal~y_exp", data = df1).fit()
```

```python
sns.regplot(x="y_exp", y='sal', data=df1)
```

```python
#coefficients
model.params
```

```python
#R-squared values
(model.rsquared, model.rsquared_adj)
```

```python
#t-value and p-value
print(model.tvalues, 'n/', model.pvalues)
```

```python
# predicting salary
data = pd.Series([3,7])
data
```

```python
data_pred = pd.DataFrame(data, columns=['y_exp'])
data_pred
```

```python
model.predict(data_pred)
```
